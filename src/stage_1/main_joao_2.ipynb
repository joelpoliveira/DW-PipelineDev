```python
df_ucars = pd.read_csv(PROC_PATH + "used_cars_processed.csv")
df_ucars.head()
df_ucars.info()
df_ucars.isna().sum()[df_ucars.isna().sum()>0]
```

```python
df_ucars.Engine.value_counts()
```

```python
#make everything lower case 
df_ucars['Engine'] = df_ucars['Engine'].str.lower()
df_ucars.Engine.value_counts()
```

##### We got more matches

```python
#if number of cylinders is in the engine name than put it on NumCylinders

# Filter rows where NumCylinders is 0 and Engine contains v8...
mask = (df_ucars['NumCylinders'] == 0) & (df_ucars['Engine'].str.contains('v8|v7|v6|v5|v4|v3|v2|v1|i8|i7|i6|i5|i4|i3|i2|i1'))

# Set NumCylinders based on Engine
df_ucars.loc[mask & df_ucars['Engine'].str.contains('v8|i8'), 'NumCylinders'] = 8
df_ucars.loc[mask & df_ucars['Engine'].str.contains('v7|i7'), 'NumCylinders'] = 7
df_ucars.loc[mask & df_ucars['Engine'].str.contains('v6|i6'), 'NumCylinders'] = 6
df_ucars.loc[mask & df_ucars['Engine'].str.contains('v5|i5'), 'NumCylinders'] = 5
df_ucars.loc[mask & df_ucars['Engine'].str.contains('v4|i4'), 'NumCylinders'] = 4
df_ucars.loc[mask & df_ucars['Engine'].str.contains('v3|i3'), 'NumCylinders'] = 3
df_ucars.loc[mask & df_ucars['Engine'].str.contains('v2|i2'), 'NumCylinders'] = 2
df_ucars.loc[mask & df_ucars['Engine'].str.contains('v1|i1'), 'NumCylinders'] = 1
#pd.set_option('display.max_rows', 20) #show more rows
df_ucars.head(20)

```
